```python
q1=np.percentile(data['male'],25, interpolation='midpoint')
q2=np.percentile(data['male'],50, interpolation='midpoint')
q3=np.percentile(data['male'],75, interpolation='midpoint')
print(q1)
print(q2)
print(q3)
```

```python
#inter quartile range
iqr=q3-q1


low_limit=q1-1.5*iqr
up_limit=q3+1.5*iqr
print(low_limit)
print(up_limit)
```

```python
outlier=[]
for x in data['male']:
    if((x>up_limit)or(x<low_limit)):
        outlier.append(x)
        
outlier        
       
```

```python
data.columns
y=data['TenYearCHD']
x=data.drop('TenYearCHD',axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn import linear_model
lr=linear_model.LinearRegression()

model=lr.fit(x_train,y_train)
prediction=model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
print('MSE Is:',mean_squared_error(y_test,prediction))
```

```python
from sklearn.metrics import r2_score
print('R squared value iis:',r2_score(y_test,prediction))
```
