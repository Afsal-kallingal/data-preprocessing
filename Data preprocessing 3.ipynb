```python
plt.boxplot(data['High'])
plt.title('High')
```

```python
q1=np.percentile(data['High'],25, interpolation='midpoint')
q2=np.percentile(data['High'],50, interpolation='midpoint')
q3=np.percentile(data['High'],75, interpolation='midpoint')
print(q1)
print(q2)
print(q3)
```

```python
data['High'].median()
```

```python
iqr=q3-q1


low_limit=q1-1.5*iqr
up_limit=q3+1.5*iqr
print(low_limit)
print(up_limit)
```

```python
outlier=[]
for x in data['High']:
    if((x>up_limit)or(x<low_limit)):
        outlier.append(x)
        
outlier        
```

```python
ind1=data['High']>up_limit
a=data.loc[ind1].index
a
```

```python
for i in a:
    data.drop([i],inplace=True)
```

```python
plt.boxplot(data['High'])
plt.title('High')
```

```python
data.head()
```

```python
data=data.drop(['Date'],axis=1)
```

```python
data.columns
y=data['Volume']
x=data.drop('Volume',axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python

from sklearn import linear_model
lr=linear_model.LinearRegression()

model=lr.fit(x_train,y_train)
prediction=model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
print('MSE Is:',mean_squared_error(y_test,prediction))
```

```python
from sklearn.metrics import r2_score
print('R squared value iis:',r2_score(y_test,prediction))
```
